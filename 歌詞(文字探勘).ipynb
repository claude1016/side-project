```python
import jieba
from jieba import analyse
import pandas as pd
from wordcloud import WordCloud
```

```python
jieba.set_dictionary('jieba_big_chinese_dict/dict.txt.big')
```

```python
content=open('窮到發慌.txt','r',encoding='utf-8').read()
```

```python
content
```

```python
words=jieba.lcut(content)
```

```python
words
```

```python
result=[]
for word in words:
    if word!='\ufeff' and word!= '\n' and  word!=' 'and  word!=')' and word!='(' : 
        result.append(word)
print(result)
```

```python
words_1=','.join(result)
```

```python

keys=jieba.analyse.extract_tags(words_1,10)
```

```python
keys
```

```python
TFIDF=jieba.analyse.extract_tags(words_1,31,withWeight=True,allowPOS=('n'))
```

```python
TFIDF
```

```python
from collections import Counter
result_1=[]
def word_frequency(text):
       
        words=[word for word in jieba.lcut(text) 
           if len(word) >= 2]
        c = Counter(words)
        for word_freq in c.most_common(31):
            result_1.append(word_freq)
        
        
```

```python
 word_frequency(words_1)
```

```python
result_1
```

```python
bad_1=pd.DataFrame(result_1,columns=['詞','出現次數'])
```

```python
bad_1
```

```python
bad=pd.DataFrame(TFIDF,columns=['名詞關鍵字','權重',])
bad['詞','出現次數']=result_1
```
