```python
import requests
from bs4 import BeautifulSoup as bf
import pandas as pd

print('蘋果火線話題')
domain='https://tw.appledaily.com'
dom = requests.get('https://tw.appledaily.com/column/index').text  # 取得蘋果今日焦點網頁
soup = bf(dom, 'html5lib')  # 將網頁丟入BeautifulSoup處理
data_1=[]
for ele in soup.find('ul',{'class':'list red'}).find_all('a') :#找出標題
             data_1.append( ele.text)

print(len(data_1))    
```

```python

data=[]
for news in soup.select('.aht_title a'):
            dom_1= requests.get(domain + news['href']).text #取得連結網頁
            soup_1= bf(dom_1, 'html5lib')
            data.append(soup_1.select_one('#bcontent').text.strip())
print(len(data))
```

```python
Big_Data={'title':data_1,'content':data}
frame=pd.DataFrame(Big_Data)
print(frame)
```
